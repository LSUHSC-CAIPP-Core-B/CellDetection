# Gen images and masks for U - Net
## Imports

```python
import cv2
import numpy as np
import os

from config import *
from CellProcessor import read_image, process_image, get_bboxes, crop_img_from_label, read_yolo_labels

LABELED_IMAGES_DIR_PATH = IMAGES_PATH + "Labels_phase_aug/"
IMAGES_MASKS_DIR_PATH = IMAGES_PATH + "Masks_phase_aug/"
IMAGES_PHASE_PATH = IMAGES_PATH + "TSV_phase_aug/"

UNET_IMAGES_PHASE_PATH = IMAGES_PATH + "unet_phase_2/"
UNET_IMAGES_MASKS_PATH = IMAGES_PATH + "unet_masks_2/"
```

## For each image

```python
labels = sorted(os.listdir(LABELED_IMAGES_DIR_PATH))
len_labels = len(labels)
curr_label_num = 1

for label in labels:
    if label.endswith(".txt") and label != "classes.txt":
        #print(label)
        print(str(curr_label_num) + "/" + str(len_labels))
        curr_label_num+=1

        image = label.split(".")[0] + ".png"
        base_image_path = IMAGES_PHASE_PATH + image
        mask_image_path = IMAGES_MASKS_DIR_PATH + image
        img_base = cv2.imread(base_image_path, cv2.IMREAD_UNCHANGED)
        img_mask = cv2.imread(mask_image_path, cv2.IMREAD_UNCHANGED)

        label_file = LABELED_IMAGES_DIR_PATH + label
        bboxes, classes = read_yolo_labels(label_file)
        k = 0
        for li in range(len(bboxes)):
            bbox = bboxes[li]
            #print(bbox)
            img_crop, img_mask = crop_img_from_label(img_base, bbox, padding=12, retCords=False, withMask=True, img_mask=img_mask)

            name_img = UNET_IMAGES_PHASE_PATH + label.split(".")[0] + "_" + str(k) + ".png"
            name_mask = UNET_IMAGES_MASKS_PATH + label.split(".")[0] + "_" + str(k) + ".png"
            k+=1
            img_crop = cv2.resize(img_crop, (128,128))
            img_mask = cv2.resize(img_mask, (128,128))
            cv2.imwrite(name_img, img_crop)
            cv2.imwrite(name_mask, img_mask)
```
